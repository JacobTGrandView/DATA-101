```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt; plt.style.use('ggplot')

Index = pd.read_csv('Index.csv')
Index.head()
```

```python
Index['Gender'].value_counts().plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('Frequency')
```

```python
Index['Height'].hist()
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
```

```python
plt.plot('Height', 'Weight', 'o', data = Index)
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
```
