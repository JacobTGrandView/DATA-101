```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

autos = pd.read_csv('Automobile_data.csv')
autos.head()
```

```python
plt.plot('horsepower', 'price', 'o', data = autos)
plt.xlabel('horsepower')
plt.ylabel('price')
```

```python
#Building linear model
lm_md = smf.ols(formula = 'price ~ horsepower', data = autos).fit()

#Extracting model results
lm_md.summary()
```

```python
#Compute residuals
autos['residuals'] = autos['price'] - lm_md.fittedvalues
autos.head()
```

```python
#Checking linearity
plt.plot('horsepower', 'residuals', 'o', data = autos)
plt.xlabel("Horsepower")
plt.ylabel("residuals")
plt.grid()
```

```python
#From the above, the residuals are increasing in absolute value from left to right,
#which indicates the assumption of linearity is not met
```
