```python
import pandas as pd
import numpy as np

import statsmodels.formula.api as smf

churn = pd.read_csv('Customer_Churn.csv')
churn.head()
```

```python
churn['Churn_numb'] = np.where(churn['Churn'] == 'Yes', 1, 0)
churn.head()
```

```python
#Logistic regression model
logit_md = smf.logit(formula = 'Churn_numb ~ tenure + MonthlyCharges', data = churn).fit()

#Extract model results
logit_md.summary()
```

```python
new_obs = pd.DataFrame({'tenure': [20],
                        'MonthlyCharges': [80]})
new_obs
```

```python
#Estimate likelihood of churn
logit_md.predict(new_obs)
```
