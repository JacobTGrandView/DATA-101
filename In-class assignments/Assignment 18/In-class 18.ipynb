```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, recall_score

churn = pd.read_csv('Customer_Churn(8).csv')
churn.head()
```

```python
churn['Churn_numb'] = churn['Churn'].map({'No': 0, 'Yes': 1})
churn.head()
```

```python
X = churn[['SeniorCitizen', 'tenure', 'MonthlyCharges']]
Y = churn['Churn_numb']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y)
```

```python
logit_md = LogisticRegression().fit(X_train, Y_train)

#Predict
logit_pred = logit_md.predict_proba(X_test)[:, 1]

#Likelihood to labels
logit_label = np.where(logit_pred < .35, 0, 1)

print(f"The Logistic Regression accuracy is {accuracy_score(Y_test, logit_label)}")
print(f"The Logistic Regression recall is {recall_score(Y_test, logit_label)}")
```

```python
RF_md = RandomForestClassifier().fit(X_train, Y_train)

#Predict
RF_pred = RF_md.predict_proba(X_test)[:, 1]

#Likelihood to labels
RF_label = np.where(RF_pred < .35, 0, 1)

print(f"The Random Forest accuracy is {accuracy_score(Y_test, RF_label)}")
print(f"The Random Forest recall is {recall_score(Y_test, RF_label)}")
```

```python
GB_md = GradientBoostingClassifier().fit(X_train, Y_train)

#Predict
GB_pred = GB_md.predict_proba(X_test)[:, 1]

#Likelihood to labels
GB_label = np.where(GB_pred < .35, 0, 1)

print(f"The Gradient Boosting accuracy is {accuracy_score(Y_test, GB_label)}")
print(f"The Gradient Boosting recall is {recall_score(Y_test, GB_label)}")
```

```python
#From the above, I would use Gradient Boosting model to predict churn since the accuracy and recall is higher
```
