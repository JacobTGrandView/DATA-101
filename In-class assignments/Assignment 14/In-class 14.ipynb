```python
#4a - Using the pandas library, read the csv file and create a data-frame called churn

import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier

churn = pd.read_csv('Customer_Churn(2).csv')
churn.head()
```

```python
churn['Churn_numb'] = churn['Churn'].map({'Yes': 1, 'No': 0})
churn.head()
```

```python
X = churn[['tenure', 'MonthlyCharges']]
Y = churn['Churn_numb']

#Decision tree model
tree_md = DecisionTreeClassifier().fit(X, Y)

#New observation
new_obs = pd.DataFrame({'tenure': [20], 'MonthlyCharges': [80]})
new_obs
```

```python
#Predict likelihood of churn
tree_md.predict_proba(new_obs)
```
