```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf

autos = pd.read_csv('Automobile_data.csv')
autos.head()
```

```python
lm_md = smf.ols(formula = 'price ~ horsepower + length', data = autos).fit()

lm_md.summary()
```

```python
#Creating new observation
new_obs = pd.DataFrame({'horsepower': [150],
                        'length': [170]})
new_obs
```

```python
#Predicting car price
lm_md.predict(new_obs)
```
