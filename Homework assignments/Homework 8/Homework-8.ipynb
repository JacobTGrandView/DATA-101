<h2>Exercise 5 - Use Default.csv datafile below</h2>

```python
#5a - Using the pandas library, read the csv file and called credit_default

import pandas as pd
import numpy as np

credit_default = pd.read_csv('Default.csv')
credit_default.head()
```

```python
#5b - create new variable in credit_default data-frame called default_numb that takes the value of 1 when 
#default = Yes and 0 when default = No.

credit_default['default_numb'] = np.where(credit_default['default'] == 'Yes', 0, 1)
credit_default.head()
```

```python
#5c - Build random forest classifier model where default_numb target variable, balance and income are input.

from sklearn.ensemble import RandomForestClassifier

X = credit_default[['balance', 'income']]
Y = credit_default['default_numb']

RF_md = RandomForestClassifier().fit(X, Y)
```

```python
#5d - predict the likelihood of default of a customer with balance = $2,000 and income = $70,000.

obsv = pd.DataFrame({'balance':[2000], 'income':[70000]})

RF_pred = RF_md.predict_proba(obsv)
print(RF_pred)
```

```python
#5e -  create a new variables in the credit_default data-frame called student_numb that takes the value of 1 when 
#student = Yes and 0 when student = No.

credit_default['student_numb'] = np.where(credit_default['student'] == 'Yes', 0, 1)
credit_default.head()
```

```python
#5f - Build random forest classifier model where default_numb is target. balance, income and student_numb are input.

X1 = credit_default[['balance', 'income', 'student_numb']]
Y1 = credit_default['default_numb']

RF_md1 = RandomForestClassifier().fit(X1, Y1)
```

```python
#5g - predict likelihood of default of customer with balance = $2,500 and income = $20,000 and student_numb = 1.

obsv1 = pd.DataFrame({'balance':[2500], 'income':[20000], 'student_numb':[1]})

RF_pred1 = RF_md1.predict_proba(obsv1)
print(RF_pred1)
```

<h2>Exercise 6 - Use Churn_Data.csv datafile below</h2>

```python
#6a - Using the pandas library, read the csv file and create a data-frame called customer_churn.

customer_churn = pd.read_csv('Churn_Data.csv')
customer_churn.head()
```

```python
#6b - create a variable called Churn_numb that takes the value of 1 when Churn = Yes and 0 when Churn = No.

customer_churn['Churn_numb'] = np.where(customer_churn['Churn'] == 'Yes', 0, 1)
customer_churn.head()
```

```python
#6c - Build a random forest classifier, in which Churn_numb is the target variable. Age, EstimatedSalary, 
#CreditScore, Balance, and NumOfProducts are the input variables. Use n_estimators = 500 and max_depth = 3

X2 = customer_churn[['Age', 'EstimatedSalary', 'CreditScore', 'Balance', 'NumOfProducts']]
Y2 = customer_churn['Churn_numb']

RF_md2 = RandomForestClassifier(n_estimators = 500,
                                max_depth = 3).fit(X2, Y2)
```

```python
#6d - predict likelihood of churn of a customer with Age = 50, EstimatedSalary = $100,000, CreditScore = 600, 
#Balance = $100,000, and NumOfProducts = 2.

obsv2 = pd.DataFrame({'Age':[50], 'EstimatedSalary':[100000], 'CreditScore':[600], 'Balance':[100000], 'NumOfProducts':[2]})

RF_pred2 = RF_md2.predict_proba(obsv2)
print(RF_pred2)
```
